# バルスの瞬間，ホントのところは一秒に何ツイートぐらいされているの？
ツイッターをクロールしたものを適当にパースして見てみるだけの簡単なお仕事

```python
raw_file = open("resources/barusu.txt", mode="r", encoding="utf-8")
```

```python
raw_tweets = []

with raw_file:
    raw_tweets = [l for l in raw_file]
```

```python
raw_tweets[0:5] # タブ区切りだった
```

```python
tweets = list()
times = list()

for t in raw_tweets:
    tmp = t.strip().split("\t")
    if len(tmp) == 2:
        tweets.append(tmp[0])
        times.append(tmp[1])
```

```python
from datetime import datetime
times = [datetime.strptime(time, "%Y-%m-%d %H:%M:%S") for time in times]
```

```python
import pandas
```

```python
df = pandas.DataFrame({
    "tweet" : tweets,
    "time" : times
})
```
